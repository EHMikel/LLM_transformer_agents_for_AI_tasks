Para el ID de imagen necesitamos añadir 2133 a las imágenes de la segunda parte ya que la última imagen de la primera parte tiene el id 2133. Para la columna ID de la segunda parte del conjunto de datos en los archivos de prueba/entrenamiento, tenemos que añadir 4737 ya que es el mayor ID entre los dos archivos de la primera parte del conjunto de datos.

```python
train_df.loc[train_df.index > 4309, 'image_id'] += 2133
train_df.loc[train_df.index > 4309, 'id'] += 4737
train_df.iloc[4307:4316]
```

```python
test_df.loc[test_df.index > 809, 'image_id'] += 2133
test_df.loc[test_df.index > 809, 'id'] += 4737
test_df.iloc[806:820]
```

```python
mapping.loc[mapping.index > 2132, 'id'] += 2133
mapping.iloc[2130:2140]
```

Parece que hemos solucionado el problema. He visto que puede haber otro problema de fuga de datos de entrenamiento (datos duplicados en entrenamiento/prueba). Vamos a comprobar si existe y tratar de solucionarlo

```python
testIDS = set(test_df['id'])
trainIDS = set(train_df['id'])
duplicated = testIDS.intersection(trainIDS)
len(duplicated)
```

```python
min(duplicated)
```

```python
train_df.loc[train_df['id'] == 26, :]
```

```python
test_df.loc[test_df['id'] == 26, :]
```

Efectivamente, aquí hay duplicación de datos.Veo dos maneras de resolver esto: eliminar algunos datos del conjunto de entrenamiento/prueba, o fusionar dos archivos sin duplicación para que uno pueda reutilizarlos y hacer una división a su elección.Yo me quedaría con la segunda opción. Debería haber:

len(test) + len(train) - len(duplicated) = 1666 + 7549 - 1193 = 8022 total rows

```python
data = pd.concat([train_df, test_df]).drop_duplicates().sort_values(['id'])
```
